```python
def dLabelCond(df, a_col, label, cond):
    cols= [a_col, label]
    L = df.loc[:, cols].copy() # nuevo dataframe con las columnas que nos importan(a_col, label)
    #print("El df L es: ")# Para imprimir el nuevo dataframe que copiamos
    #print(L)
    newDf = L.loc[L[label]==cond].copy() #generamos un nuevo dataframe con solo la opcion de 'cond' en 'label'
    #print("El df newDf es: ")
    #print(newDf)
    #print(newDf[a_col].size) #imprime la cantidad de rows
    return newDf
    #print(L.value_counts())
#pTenis(dataTenis, 'Humidity','Play Tennis','Yes')
#pTenis(dataTenis, 'Humidity', 'Play Tennis','No')
```

```python
def pColDueToLabel(dataframe, col, condCol, label, condLabel):
    firstResult = dLabelCond(dataframe, col, label, condLabel)
    newDf = firstResult.loc[firstResult[col]==condCol].copy()
    X= firstResult.shape[0]#Cantidad de condUno para la columna 'Play Tennis que es la objetivo'
    Y= newDf.shape[0] #
    totalRows = dataframe.shape[0] #
    #print("X es", X)
    if (X):
        result= Y/X * X/totalRows
        #print('El resultado es:', result)
        return result
    
    #print('X es:', X)#Cantidad de 'condLabel' en la columna 'label'
    #print('Y es:', Y)#Cantidad de 'condCol' en la columna 'col' respecto a las filas que son 'condLabel' en la columna 'label'
    #print('totalRows es:', totalRows)#Cantidad total de columnas en el conjunto de datos
    #print('Dado que X es 0, no hay probabilidad sin embargo se devolvera 0.0001')#Resultado al que se concluye de una columna respecto a la otra
    result = Y/1 * 1/totalRows
        #print('El resultado es:', result)
    return result
```

```python
def setM1(fila, df):
    first_prob=pColDueToLabel(df, 'Cupo2', 'Cupo_3', 'Marca2', 'Marca_20' )
    second_prob=pColDueToLabel(df, 'CapacidadEnvase2', 'CapacidadEnvase_9', 'Marca2', 'Marca_20' )
    prob_total= first_prob * second_prob
    datasetthree.at[fila,"Marca1"]= prob_total
    #datasetthree.head()
def setM2(fila, df):
    first_prob=pColDueToLabel(df, 'Cupo2', 'Cupo_2', 'Marca2', 'Marca_16' )
    second_prob=pColDueToLabel(df, 'CapacidadEnvase2', 'CapacidadEnvase_10', 'Marca2', 'Marca_16' )
    prob_total= first_prob * second_prob
    datasetthree.at[fila,"Marca2"]= prob_total
    #datasetthree.head()
def setM3(fila, df):
    first_prob=pColDueToLabel(df, 'Cupo2', 'Cupo_3', 'Marca2', 'Marca_9' )
    second_prob=pColDueToLabel(df, 'CapacidadEnvase2', 'CapacidadEnvase_12', 'Marca2', 'Marca_9' )
    prob_total= first_prob * second_prob
    datasetthree.at[fila,"Marca3"]= prob_total
    #datasetthree.head()
def setMI1(fila, df):
    first_prob=pColDueToLabel(df, 'Cupo2', 'Cupo_2', 'Marca2', 'Marca_38' )
    second_prob=pColDueToLabel(df, 'CapacidadEnvase2', 'CapacidadEnvase_10', 'Marca2', 'Marca_38' )
    prob_total= first_prob * second_prob
    datasetthree.at[fila,"Marca_Inno1"]= prob_total
    #datasetthree.head()
def setMI2(fila, df):
    first_prob=pColDueToLabel(df, 'Cupo2', 'Cupo_2', 'Marca2', 'Marca_39' )
    second_prob=pColDueToLabel(df, 'CapacidadEnvase2', 'CapacidadEnvase_10', 'Marca2', 'Marca_39' )
    prob_total= first_prob * second_prob
    datasetthree.at[fila,"Marca_Inno2"]= prob_total
    #datasetthree.head()
```

```python
def iterarFilas(x):
    for index, row in datasetthree.iterrows():
        x = datasettwo[datasettwo['Cliente'] == row['Cliente']].copy()
        setM1(index, x)
        setM2(index, x)
        setM3(index, x)
        setMI1(index, x)
        setMI2(index, x)
        #print("Va en la fila: ", index)
        #datasetthree.head()
```
